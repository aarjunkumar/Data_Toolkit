Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap

N = 30
x = np.random.rand(N) * 100
y = np.random.rand(N) * 100

sizes = np.random.randint(10, 30, size=N)
colors = np.random.rand(N) * 100

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Scatter plot",x_axis_label='X', y_axis_label='Y')

m = linear_cmap(field_name='colors', palette='Viridis256', low=min(colors), high=max(colors))

p.scatter('x', 'y', size='sizes', color=m, source=source, fill_alpha=0.6)

show(p)
```

Q3.  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

  fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

  counts = [20, 25, 30, 35]

```python
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits,title="Fruit Counts")

p.vbar(x='fruits', top='counts', width=0.7, source=source, fill_color=factor_cmap('fruits', palette=['red', 'orange', 'yellow', 'green'], factors=fruits))

show(p)
```

Q4. Create a Bokeh histogram to visualize the distribution of the given data.

  data_hist = np.random.randn(1000)
  
  hist, edges = np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure,show,output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

source = ColumnDataSource(data=dict(
    top=hist,
    left=edges[:-1],
    right=edges[1:]
))

p=figure()

p.quad(top='top', bottom=0, left='left', right='right',
       source=source,alpha=0.5)

show(p)
```

Q5. Create a Bokeh heatmap using the provided dataset.


 data_heatmap = np.random.rand(10, 10)

 x = np.linspace(0, 1, 10)

 y = np.linspace(0, 1, 10)

 xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show, output_file
from bokeh.models import LinearColorMapper, ColorBar, ColumnDataSource
from bokeh.transform import transform
from bokeh.io import output_notebook

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

source = ColumnDataSource(data=dict(x=xx.flatten(),y=yy.flatten(),z=data_heatmap.flatten()))

c_mapper=LinearColorMapper(palette='Viridis256', low=min(data_heatmap.flatten()), high=max(data_heatmap.flatten()))

p=figure(title='Heatmap', x_range=(0,1), y_range=(0,1))

p.rect(x='x',y='y',width=0.1, height=0.1,fill_color=transform('z', c_mapper), line_color=None, source=source)

color_bar=ColorBar(color_mapper=c_mapper,label_standoff=12)
p.add_layout(color_bar,'right')

show(p)
```
