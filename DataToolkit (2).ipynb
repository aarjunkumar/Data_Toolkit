Q3.  Explain the following terms

a) The difference in np.array, np.asarray and np.asanyarray

sol:
* np.array:- np.array() will always create a new array, copying the input.

* np.asarray():- np.asarray() converts the input to a NumPy array but avoids copying if it's already an array.

* np.asanyarray():- np.asanyarray() works like np.asarray() but preserves ndarray subclasses like matrix.



```python
arr1=np.array([1,2,3])
arr2=np.asarray([1,2,3])
arr3=np.asanyarray([1,2,3])

arr1,arr2,arr3
```

b) The difference between Deep copy and shallow copy

Sol. Shallow Copy:- Creates a copy of the top-level structure of an object or array, but the nested elements are shared references with the original. Changes made to the copied object will also reflect in the original object.

Deep Copy:- Creates a completely new object or array and recursively copies all nested objects and arrays. There are no shared references with the original object, so changes made to the copied object will not affect the original object

Difference:-

The main difference between a shallow copy and a deep copy is that a shallow copy shares references with the original object, while a deep copy is a completely independent copy.



```python
x=np.array([1,2,3,4,5])

# Shallow copy

y=x
x[0]=100
print("--- Shallow copy---")
print("Original Array: ", x)
print("shallow Array: ", y)

# Deep copy

a=np.array([1,2,3,4,5])

b=a.copy()
a[0]=100

print("--- Deep copy---")
print("Original Array: ", a)
print("shallow Array: ", b)
```

Q13. Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the
following tasks using Matplotlib and NumPy:

 a) Create a scatter plot using x and y, setting the color of the points to red and the marker style to 'o'.

 b) Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.

 c) Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.

 d) Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'.

Q15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:

 a) Plot a histogram of the data with 30 bins.

 b) Overlay a line plot representing the normal distribution's probability density function (PDF).

 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

 d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
from scipy.stats import norm

x=np.random.randn(1000)

plt.hist(x,bins=30,density=True,edgecolor='white',label='Histogram')
plt.xlabel("Value")
plt.ylabel("Frequency/Probability")
plt.title("Histogram with PDF Overlay")

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)

plt.plot(x,p,'k',linewidth=1,label="Normal PDF")

plt.legend()
plt.show()
```

Q17.  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
x = np.random.randn(100)
y = np.random.randn(100)

def quad(x,y):
  if x>=0 and y>=0:
    return 'Q1'
  elif x<0 and y>0:
    return 'Q2'
  elif x<0 and y<0:
    return 'Q3'
  else:
    return 'Q4'

df=pd.DataFrame({'x':x,'y':y,'quadrant':list(map(quad,x,y))})

sns.scatterplot(data=df,x='x',y='y',hue='quadrant')
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Quadrant-wise Scatter Plot")
plt.show()
```

Q19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Bright5

output_notebook()

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)

source=ColumnDataSource(data=dict(categories=categories,values=values,color=Bright5))

p=figure(x_range=categories,title="Random Categorical Bar Chart",y_range=(1,100),tools="",toolbar_location=None)

p.vbar(x='categories',top='values',color='color',source=source,width=0.9)

p.xaxis.axis_label='Categories'
p.yaxis.axis_label="Values"

hover=HoverTool()
hover.tooltips=[('Category','@categories'),("Value",'@values')]
p.add_tools(hover)

show(p)
```

Q20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
